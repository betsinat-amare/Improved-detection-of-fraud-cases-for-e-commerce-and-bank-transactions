```python
import pandas as pd
import numpy as np

fraud_df = pd.read_csv("../data/raw/Fraud_Data.csv")
```

```python
fraud_df["signup_time"] = pd.to_datetime(fraud_df["signup_time"])
fraud_df["purchase_time"] = pd.to_datetime(fraud_df["purchase_time"])
```

```python
fraud_df["hour_of_day"] = fraud_df["purchase_time"].dt.hour
fraud_df["day_of_week"] = fraud_df["purchase_time"].dt.dayofweek

fraud_df["time_since_signup"] = (
    fraud_df["purchase_time"] - fraud_df["signup_time"]
).dt.total_seconds() / 3600
```

```python
fraud_df = fraud_df.sort_values(["user_id", "purchase_time"])

fraud_df["user_txn_count"] = fraud_df.groupby("user_id").cumcount() + 1
fraud_df["transactions_per_device"] = fraud_df.groupby("device_id")["device_id"].transform("count")
```

```python
fraud_df = fraud_df.drop(
    columns=["signup_time", "purchase_time", "ip_address"]
)
```

```python
fraud_df.to_csv("../data/processed/fraud_feature_engineered.csv", index=False)
```
